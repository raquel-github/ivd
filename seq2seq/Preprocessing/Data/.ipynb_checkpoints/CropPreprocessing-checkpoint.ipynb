```python
from PIL import Image
import json
import shutil
import os 
import glob
```

```python
tr_file = '../../../../data/guesswhat.test.new.jsonl'
# tr_file = 'guesswhat.small.test.jsonl'
with open(tr_file) as f:
    lines = f.readlines()
    data_training = [json.loads(x.strip()) for x in lines]
print(len(data_training))
data_training[15]
```

```python
for datum in data_training:
    target_id = datum['object_id']
    img_file = datum['image']['file_name']
    img = Image.open('../../../../data/GuessWhat/Test/'+img_file)
    objects = datum['objects']
    for obj in objects:
        if target_id == obj['id']:
            bbox = obj['bbox']
            x = bbox[0]
            y = bbox[1]
            w = bbox[2]
            h = bbox[3]

            area = (int(x), int(y), int(x + w), int(y + h))
            cropped_img = img.crop(area)
            cropped_img.save('../../../../data/GuessWhat/Test_Crops/'+str(datum['id'])+'.jpg')
```

```python
target_id = data_training[0]['object_id']
img_file = data_training[0]['image']['file_name']
img = Image.open('../../../../data/GuessWhat/Validation/'+img_file)
img.show()
objects = data_training[0]['objects']
print(target_id)
for obj in objects:
    if target_id == obj['id']:
        bbox = obj['bbox']
        x = bbox[0]
        y = bbox[1]
        w = bbox[2]
        h = bbox[3]
        
        area = (int(x), int(y), int(x + w), int(y + h))
        cropped_img = img.crop(area)
        cropped_img.show()
#         cropped_img.save(str(data_training[0]['id'])+'.jpg')
```
