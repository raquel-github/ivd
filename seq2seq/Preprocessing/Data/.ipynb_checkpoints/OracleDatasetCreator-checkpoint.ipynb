```python
def img_spatial(img_meta):
    """ returns the spatial information of a bounding box """
    bbox            = img_meta[0] # gets all bboxes in the image

    width           = img_meta[1]
    height          = img_meta[2]

    image_center_x  = width / 2
    image_center_y  = height / 2
    
    x_min = bbox[0] / width
    y_min = bbox[1] / height
    x_max = (bbox[0] + bbox[2]) / width
    y_max = (bbox[1] + bbox[3]) / height


    w_box = x_max - x_min
    h_box = y_max - y_min

    x_min = x_min * 2 - 1
    y_min = y_min * 2 - 1
    x_max = x_max * 2 - 1
    y_max = y_max * 2 - 1


    x_center = (x_min + x_max) / 2
    y_center = (y_min + y_max) / 2


    spatial = [Decimal(format(x_min, '.4f')), Decimal(format(y_min, '.4f')), Decimal(format(x_max, '.4f')),\
               Decimal(format(y_max, '.4f')), Decimal(format(x_center, '.4f')), Decimal(format(y_center, '.4f')),\
               Decimal(format(w_box, '.4f')), Decimal(format(h_box, '.4f'))]


    return spatial
```

```python
questions = []
for datum in data_training:
    target_id = datum['object_id']
    game_id = datum['id']
    img_file = datum['image']['file_name']
    objects = datum['objects']
    for obj in objects:
        if target_id == obj['id']:
            bbox = obj['bbox']
            obj_cat = obj['category_id']
    img_meta = [bbox, datum['image']['width'], datum['image']['height']]
    spatial = (img_spatial(img_meta))
    qas = datum['qas']
    
    for qa in qas:
        questions.append({'answer':qa['answer'], 'question':qa['question'], 'spatial':spatial, 'img_features':img_file, \
                          'crop_features':str(game_id)+'.jpg', 'obj_cat':obj_cat, 'game_id':game_id})
```

```python
json_data = {'questions':questions}

with open('oracle.train.json', 'w') as file:
    simplejson.dump(json_data, file)
```
