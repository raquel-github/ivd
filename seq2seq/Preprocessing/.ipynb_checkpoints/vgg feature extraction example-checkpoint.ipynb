```python
import vgg
import io
import requests
from PIL import Image
import torch
from torchvision import models, transforms
from torch.autograd import Variable
import torch.nn as nn
```

```python
model = vgg.vgg16_bn(pretrained=True)
model.classifier = nn.Sequential(
    nn.Linear(512 *7 *7, 4096),
    nn.ReLU(),
    nn.Dropout(),
    nn.Linear(4096, 4096)
)
```

```python
normalize = transforms.Normalize(
   mean=[0.485, 0.456, 0.406],
   std=[0.229, 0.224, 0.225]
)
preprocess = transforms.Compose([
   transforms.Scale(torch.Tensor([224,224])),
   #transforms.CenterCrop(224),
   transforms.ToTensor(),
   normalize
])
```

```python
response = requests.get('https://s3.amazonaws.com/outcome-blog/wp-content/uploads/2017/02/25192225/cat.jpg')
img_pil = Image.open(io.BytesIO(response.content))
print(img_pil.size)
```

```python
import numpy
img_p = 'img/GuesserModel.png'
channels = numpy.array(Image.open(img_p).convert('RGB').getdata(), dtype=numpy.float)
print(channels.shape)
print(numpy.mean(channels[:,0]))
print(numpy.mean(channels[:,1]))
print(numpy.mean(channels[:,2]))
```

```python
img_tensor = preprocess(img_pil)
print(img_tensor.size())
img_tensor.unsqueeze_(0)
img_tensor
```

```python
img_variable = Variable(img_tensor)
fc_out = model(img_variable)
```
